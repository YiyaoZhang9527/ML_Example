```python
import numpy as np
import sympy
f = sympy.Function("f")
from sympy.abc import x,y,z
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
sns.set()
```

### $ 第一个例子M(t)=2^{x}次方 $
### 比较好的方法是将指数函数看做质量

```python
T = np.arange(-1,1,32)
```

```python
t = sympy.symbols('t')
f1 = 2**t
f1
```

```python
tnumber = [f1.subs(t,i).evalf(7) for i in range(0,32)]
date = [i for i in range(32)]
ppl = pd.DataFrame({'number':tnumber,'date':date})
ppl.head()
```

```python
x,y=ppl.date,ppl.number
```

```python
plt.figure(figsize=(10,8))
plt.plot(x,y)
plt.scatter(x,y, label='every day',color='r')
plt.xlabel('number')
plt.ylabel('date')
plt.title('ppl add')
plt.yticks()
plt.legend()
plt.show()
```

### $从上图可以看出，函数M(t)=2^{x}的变化会随着每一天时间的增加而倍数的增加.$
### $，但是这里我们要求的是导数啊导数是微小的变化，而不是一天，$
### $所以，这就是为什么要将函数M(t)=2^{x}看作质量的原因了,因为质量是无限分割才比较符合人的直觉$
