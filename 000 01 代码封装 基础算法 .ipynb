```python
def binarySearch(arr, left, right, x):
    while left <= right:
        mid = int(left + (right - left) / 2); # æ‰¾åˆ°ä¸­é—´ä½ç½®ã€‚æ±‚ä¸­ç‚¹å†™æˆ(left+right)/2æ›´å®¹æ˜“æº¢å‡ºï¼Œæ‰€ä»¥ä¸å»ºè®®è¿™æ ·å†™
        # æ£€æŸ¥xæ˜¯å¦å‡ºç°åœ¨ä½ç½®mid
        if arr[mid] == x:
            print('found %d åœ¨ç´¢å¼•ä½ç½®%d å¤„' %(x,mid))
            return mid
            # å‡å¦‚xæ›´å¤§ï¼Œåˆ™ä¸å¯èƒ½å‡ºç°åœ¨å·¦åŠéƒ¨åˆ†
        elif arr[mid] < x:
            left = mid + 1 #æœç´¢åŒºé—´å˜ä¸º[mid+1,right]
            print('åŒºé—´ç¼©å°ä¸º[%d,%d]' %(mid+1,right))
        elif x<arr[mid]:
            right = mid - 1 #æœç´¢åŒºé—´å˜ä¸º[left,mid-1]
            print('åŒºé—´ç¼©å°ä¸º[%d,%d]' %(left,mid-1))
    return -1
```

```python
def search_n(s, c, n):
    size = 0
    for i, x in enumerate(s):
        if x == c:
            size += 1
        if size == n:
            return i
    return -1
print(search_n("fdasadfadf", "a", 3))# ç»“æœä¸º7ï¼Œæ­£ç¡®
print(search_n("fdasadfadf", "a", 30))# ç»“æœä¸º-1ï¼Œæ­£ç¡®
```

```python
import operator
from collections import Counter
import sys
from math import ceil, floor
import re
from copy import deepcopy
from random import randint


# TODO 1. é‡å¤å…ƒç´ åˆ¤å®š
def all_unique(lst):
    return len(lst) == len(set(lst))


x = [1, 1, 2, 2, 3, 2, 3, 4, 5, 6]
y = [1, 2, 3, 4, 5]
print(all_unique(x))  # False
print(all_unique(y))  # True


# TODO 2. å­—ç¬¦å…ƒç´ ç»„æˆåˆ¤å®š
def anagram(first, second):
    return Counter(first) == Counter(second)


print(anagram("abcd3", "3acdb"))  # True

# TODO 3. å†…å­˜å ç”¨
variable = 30
print(sys.getsizeof(variable))  # 24


# TODO 4. å­—èŠ‚å ç”¨
def byte_size(string):
    return (len(string.encode('utf-8')))


print(byte_size('ğŸ˜€'))  # 4
print(byte_size('Hello World'))  # 11

# TODO 6. å¤§å†™ç¬¬ä¸€ä¸ªå­—æ¯
s = "programming is awesome"

print(s.title())


# TODO 7. åˆ†å— ç»™å®šå…·ä½“çš„å¤§å°ï¼Œå®šä¹‰ä¸€ä¸ªå‡½æ•°ä»¥æŒ‰ç…§è¿™ä¸ªå¤§å°åˆ‡å‰²åˆ—è¡¨ã€‚
def chunk(lst, size):
    return list(map(lambda x: lst[x * size:x * size + size], list(range(0, ceil(len(lst) / size)))))


print(chunk([1, 2, 3, 4, 5], 2))


# TODO 8. å‹ç¼© è¿™ä¸ªæ–¹æ³•å¯ä»¥å°†å¸ƒå°”å‹çš„å€¼å»æ‰ï¼Œä¾‹å¦‚ï¼ˆFalseï¼ŒNoneï¼Œ0ï¼Œâ€œâ€ï¼‰ï¼Œå®ƒä½¿ç”¨ filter() å‡½æ•°
def compact(lst):
    return list(filter(bool, lst))


compact([0, 1, False, 2, '', 3, 'a', 's', 34])

# TODO 9. è§£åŒ… å¦‚ä¸‹ä»£ç æ®µå¯ä»¥å°†æ‰“åŒ…å¥½çš„æˆå¯¹åˆ—è¡¨è§£å¼€æˆä¸¤ç»„ä¸åŒçš„å…ƒç»„ã€‚
array = [['a', 'b'], ['c', 'd'], ['e', 'f']]
transposed = zip(*array)
print(list(transposed))


# TODO 12. å…ƒéŸ³ç»Ÿè®¡
def count_vowels(str):
    return len(re.findall(r'[aeiou]', str, re.IGNORECASE))


print(count_vowels('foobar'))  # 3
print(count_vowels('gym'))  # 0


# TODO 13. é¦–å­—æ¯å°å†™
def decapitalize(string):
    return string[:1].lower() + string[1:]


print(decapitalize('FooBar'))  # 'fooBar'
print(decapitalize('FooBar'))  # 'fooBar'


# TODO 14. å±•å¼€åˆ—è¡¨ï¼ˆlisté™ç»´ï¼‰è¯¥æ–¹æ³•å°†é€šè¿‡é€’å½’çš„æ–¹å¼å°†åˆ—è¡¨çš„åµŒå¥—å±•å¼€ä¸ºå•ä¸ªåˆ—è¡¨
def spread(arg):
    '''
    å¹¿æ’­å‡½æ•°
    :param arg:
    :return:
    '''
    ret = []
    for i in arg:
        if isinstance(i, list):
            ret.extend(i)  # åˆ—è¡¨æœ«å°¾è¿½åŠ  ç­‰åŒäº ret += i
        else:
            ret.append(i)
    return ret


def deep_flatten(lst):
    '''
    æ·±åº¦å¹³å±•
    :param lst:
    :return:
    '''
    result = []
    result.extend(
        spread(list(map(lambda x: deep_flatten(x) if type(x) == list else x, lst))))
    return result


print(deep_flatten([1, [2], [[3], 4], 5, [2, [9, [10, 2, 2, 3]]]]))  # [1,2,3,4,5]


# TODO 15.1 åˆ—è¡¨çš„å·® è¯¥æ–¹æ³•å°†è¿”å›ç¬¬ä¸€ä¸ªåˆ—è¡¨çš„å…ƒç´ ï¼Œå…¶ä¸åœ¨ç¬¬äºŒä¸ªåˆ—è¡¨å†…ã€‚å¦‚æœåŒæ—¶è¦åé¦ˆç¬¬äºŒä¸ªåˆ—è¡¨ç‹¬æœ‰çš„å…ƒç´ ï¼Œè¿˜éœ€è¦åŠ ä¸€å¥ set_b.difference(set_a)ã€‚
def difference1(a, b):
    set_a = set(a)
    set_b = set(b)
    comparison = set_a.difference(set_b)
    return list(comparison)


print(difference1([1, 2, 3], [1, 2, 4]))


# TODO 15.2 åˆ—è¡¨çš„å·® ç­‰åŒäºä¸Šä¸€ä¸ªï¼Œä½†æ˜¯ç®€çŸ­å¤šäº†ï¼ŒçœŸä¸çŸ¥é“ä¸ºä»€ä¹ˆè¿˜æœ‰äººä¼šç”¨differeneceï¼Ÿï¼Ÿï¼Ÿ
def difference2(a, b):
    return list(set(a) - set(b))


print(difference2([1, 2, 3], [1, 2, 4]))


# TODO åŸºæœ¬é›†åˆè®¡ç®—å±•ç¤º
def collection_of_mathematical(a, b):
    seta, setb = set(a), set(b)
    return {
        'the b members of a': b in a
        # æˆå‘˜
        , 'intersection': list(seta & setb)
        # äº¤é›†
        , 'and set': list(seta | setb)
        # å¹¶é›†
        , 'difference': list(seta - setb)
        # å·®é›†
        , 'complement or symmetric difference': list(seta ^ setb)
        # å¯¹ç§°å·®é›†æˆ–è€…è¡¥é›†ï¼Œå¦‚æœé›†åˆaæ˜¯bçš„è¶…é›†ï¼Œåˆ™æœ‰è¡¥é›†
        , 'The father set a of b': seta > setb
        # åˆ¤æ–­ a æ˜¯ bçš„ çˆ¶é›†
        , 'A subset of b of a': setb < seta
        # åˆ¤æ–­ b æ˜¯ a çš„ å­é›†
    }


print(collection_of_mathematical([1, 2, 3, 4], [1, 2, 4]))


# TODO 16. é€šè¿‡å‡½æ•°å–å·® å¦‚ä¸‹æ–¹æ³•é¦–å…ˆä¼šåº”ç”¨ä¸€ä¸ªç»™å®šçš„å‡½æ•°ï¼Œç„¶åå†è¿”å›åº”ç”¨å‡½æ•°åç»“æœæœ‰å·®åˆ«çš„åˆ—è¡¨å…ƒç´ ã€‚
def difference_by(a, b, fn):
    b = set(map(fn, b))
    return [item for item in a if fn(item) not in b]


print(difference_by([2.1, 1.2], [2.3, 3.4], floor))  # [1.2]
print(difference_by([{'x': 2}, {'x': 1}], [{'x': 1}], lambda v: v['x']))


# TODO 17. é“¾å¼å‡½æ•°è°ƒç”¨ ä½ å¯ä»¥åœ¨ä¸€è¡Œä»£ç å†…è°ƒç”¨å¤šä¸ªå‡½æ•°ã€‚
def add(a, b):
    return a + b


def subtract(a, b):
    return a - b


a, b = 4, 5
print((subtract if a > b else add)(a, b))  # 9


# TODO 18. æ£€æŸ¥é‡å¤é¡¹
def has_duplicates(lst):
    return len(lst) != len(set(lst))


x = [1, 2, 3, 4, 5, 5]
y = [1, 2, 3, 4, 5]
print(has_duplicates(x))  # True
print(has_duplicates(y))  # False


# TODO 19.1 åˆå¹¶ä¸¤ä¸ªå­—å…¸
def merge_two_dicts(a, b):
    c = a.copy()  # make a copy of a
    c.update(b)  # modify keys and values of a with the ones from b
    return c


a = {'x': 1, 'y': 2}
b = {'y': 3, 'z': 4}
print(merge_two_dicts(a, b))


# TODO 19.2 åˆå¹¶ä¸¤ä¸ªå­—å…¸
def merge_dictionaries(a, b):
    return {**a, **b}


print(merge_dictionaries(a, b))


# TODO 20.1 å°†ä¸¤ä¸ªåˆ—è¡¨è½¬åŒ–ä¸ºå­—å…¸
def to_dictionary1(keys, values):
    return dict(zip(keys, values))


# TODO 20.2 å°†ä¸¤ä¸ªåˆ—è¡¨è½¬åŒ–ä¸ºå­—å…¸
def to_dictionary2(keys, values):
    return {k: v for k, v in zip(keys, values)}


keys = ["a", "b", "c"]
values = [2, 3, 4]
print(to_dictionary1(keys, values))
print(to_dictionary2(keys, values))

# TODO 21. ä½¿ç”¨æšä¸¾
list_ = ["a", "b", "c", "d"]
for index, element in enumerate(list_):
    print("Value", element, "Index ", index, )


# TODO 24. å…ƒç´ é¢‘ç‡(åˆ—è¡¨ä¼—æ•°)
def most_frequent(list):
    return max(set(list), key=list.count)


list_ = [1, 2, 1, 2, 3, 2, 1, 4, 2]
print(most_frequent(list_))


# TODO 25.å›æ–‡åºåˆ—
def palindrome(string):
    '''
    ä»¥ä¸‹æ–¹æ³•ä¼šæ£€æŸ¥ç»™å®šçš„å­—ç¬¦ä¸²æ˜¯ä¸æ˜¯å›æ–‡åºåˆ—ï¼Œ
    å®ƒé¦–å…ˆä¼šæŠŠæ‰€æœ‰å­—æ¯è½¬åŒ–ä¸ºå°å†™ï¼Œå¹¶ç§»é™¤éè‹±
    æ–‡å­—æ¯ç¬¦å·ã€‚æœ€åï¼Œå®ƒä¼šå¯¹æ¯”å­—ç¬¦ä¸²ä¸åå‘å­—
    ç¬¦ä¸²æ˜¯å¦ç›¸ç­‰ï¼Œç›¸ç­‰åˆ™è¡¨ç¤ºä¸ºå›æ–‡åºåˆ—ã€‚
    :param string:
    :return:
    '''
    from re import sub
    s = sub('[\W_]', '', string.lower())
    return s == s[::-1]


print(palindrome('taco cat'))  # True

# TODO 26. ä¸ä½¿ç”¨ if-else çš„è®¡ç®—å­
action = {
    "+": operator.add,
    "-": operator.sub,
    "/": operator.truediv,
    "*": operator.mul,
    "**": pow
}

print(action['-'](50, 25))  # 25


# TODO 27.Shuffle
def shuffle(lst):
    '''
    ä¸­æ–‡åï¼šæ´—ç‰Œç®—æ³•
    è¯¥ç®—æ³•ä¼šæ‰“ä¹±åˆ—è¡¨å…ƒç´ çš„é¡ºåºï¼Œå®ƒä¸»è¦ä¼šé€šè¿‡ Fisher-Yates ç®—æ³•å¯¹æ–°åˆ—è¡¨è¿›è¡Œæ’åºï¼š
    è¯¦è§£ä½ç½®urlï¼šhttps://blog.csdn.net/qq_26399665/article/details/79831490
    :param lst:
    :return:
    '''
    temp_lst = deepcopy(lst)
    m = len(temp_lst)
    while (m):
        m -= 1
        i = randint(0, m)
        temp_lst[m], temp_lst[i] = temp_lst[i], temp_lst[m]
    return temp_lst


foo = [1, 2, 3]
print(shuffle(foo))


# TODO 29. äº¤æ¢å€¼ï¼ˆäº¤æ¢å˜é‡åä¸‹çš„å€¼ï¼‰
def swap(a, b):
    return b, a


a, b = -1, 14
print(swap(a, b))  # (14, -1)

# TODO 30. å­—å…¸é»˜è®¤å€¼
d = {'a': 1, 'b': 2}
print(d.get('c', 3))
```
