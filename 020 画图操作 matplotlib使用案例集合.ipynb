```python
import matplotlib.pyplot as plt
import numpy as np
import random

vector_x,vector_y = [random.randint(-100,100) for i in range(100)],[random.randint(-100, 100) for i in range(100)]


srcindex_x = abs(max(vector_x)) >= abs(min(vector_x)) and abs(max(vector_x)) or abs(min(vector_x))
srcindex_y = abs(max(vector_y)) >= abs(min(vector_y)) and abs(max(vector_y)) or abs(min(vector_y))
srcindex = srcindex_x >= srcindex_y and srcindex_x or srcindex_y
index = np.arange(-srcindex*1.5,srcindex*1.5)

plt.suptitle('my plot', fontsize=20, fontweight='bold')
plt.plot(index,[0 for i in range(len(index))])
plt.plot([0 for i in range(len(index))],index)
plt.scatter(vector_x,vector_y,label='root data'
                ,c='b'  # 画图颜色：c 青色 ，b 蓝色 ，r红色 ， y黄色 ，m 品红色 ，k 黑色 ，w 白色
                ,s=5
                ,marker = 'X')
plt.savefig('笛卡尔坐标系图.png')
plt.show()
plt.close()
```

```python
import random
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

print('#画动图第一张')
fig, ax = plt.subplots()
y1 = []
for i in range(50):
    y1.append(i)  # 每迭代一次，将i放入y1中画出来
    ax.cla()   # 清除键
    ax.bar(y1, label='test', height=y1, width=0.3)
    ax.legend()
    plt.pause(0.1)
plt.close()

print('第一张动图结束')

print('#画动图第二张')
def plot_():
    fig, ax = plt.subplots()
    y1 = []
    for i in range(50):
        y1.append(i)  # 每迭代一次，将i放入y1中画出来
        ax.cla()  # 清除键
        ax.plot(y1, label='test')
        ax.legend()
        plt.pause(0.1)
plt.close()

print('第二张动图结束')


print('#画动图第三张')
```
