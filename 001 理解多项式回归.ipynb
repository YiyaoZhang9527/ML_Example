```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

## 线性回归可以很好的拟合出一条线，在数据离散程度不大的时候，可以很好的帮我们拟合出一条线，用以预测价格，但是更多时候，我们可能需要的是一条曲线来拟合出更好的预测结果，这时候，多项式回归就能榜上大忙了。

```python
x = np.arange(10000)
```

## 首先理解指数函数的特性
### 当指数为2时函数图像表示如下

```python
plt.plot(x)
plt.plot(x**2)
```

### 当指数为3时，函数图像表示如下

```python
plt.plot(x)
plt.plot(x**3)
```

### 假如继续扩大指数呢？

```python
plt.plot(x**4)
```

### 但是到了指数为5，情况发生了逆转，多项式在到达5以后，曲线开始高速下沉

```python
plt.plot(x**5)
```

### 但是我们还有对数啊

```python
plt.plot(np.log2(x))
```

### 因为numpy和math库当中只有:  log2 , log10 , log(e) , log(x+1)以及log(x+1)的逆运算exp(x)-1 
### 所以我们要自己拟一个log3的函数出来，以3为底x的对数写作 : np.log(x)/np.log(3)

```python
plt.plot(np.log(x)/np.log(3))
```
