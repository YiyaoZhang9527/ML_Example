```python
import numpy as np
import matplotlib.pyplot as plt

def cross_entropy(y, t):
    delta = 1e-7
    return -np.sum(t * np.log(y + delta))

y = np.linspace(0.01, 0.99, 100)
t = np.array([1 if i == 50 else 0 for i in range(100)])
ce = np.array([cross_entropy(np.array([y_val]), np.array([t_val])) for y_val, t_val in zip(y, t)])

plt.plot(y, ce)
plt.xlabel('y')
plt.ylabel('cross-entropy')
plt.show()
```
