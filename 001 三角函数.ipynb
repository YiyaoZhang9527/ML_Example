```python
import numpy as np
import matplotlib.pyplot as plt
```

圆周长字母公式为：C=πD=2πR，π是圆周率，约等于3.14。
公式中R/r为圆的半径，D/d为圆的直径。
用文字表示为：圆的周长=直径×圆周率=半径×2×圆周率

```python
R = 2
lenC = np.linspace(0,R*np.pi,800) 
plt.figure(figsize=(10,10))
plt.plot(np.cos(lenC)*R,np.sin(lenC)*R,c = 'g')
plt.plot(np.array([-(R+1),R+1]),np.zeros(2),c='black')
plt.plot(np.zeros(2),np.array([-(R+1),R+1]),c='black')
point_in_C = np.c_[np.cos(lenC)*R,np.sin(lenC)*R][100]
plt.plot([0,point_in_C[0]],[0,point_in_C[-1]])

plt.plot([0,np.sin(1)*2],[point_in_C[-1],point_in_C[0]])
```
