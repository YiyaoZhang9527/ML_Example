## 创建样本

```python
pop_data = np.random.gamma(1,100,3000)
plt.hist(pop_data)
```

## 计算样本均值

```python
pop_data.mean()
```

## 抽样次数逐渐增加则样本量逐渐接近真实均值

```python
sum([np.random.choice(pop_data, i).mean() for i in range(1,1000)])/1000
```

# 画出抽样结果

```python
plt.plot([np.random.choice(pop_data, i).mean() for i in range(1,1000)])
```
