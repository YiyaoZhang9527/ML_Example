```python
import matplotlib.pyplot as plt
import numpy as np


def price_game(sumv=600,v=1,p=0.02):
    dv ,dp ,resultT,resultV = 0 ,0,[],[]
    for i in range(int(1/p)):
        dv += v
        dp += p
        pm = (1-dp)
        if pm > 0:
            pr = (v+dv)*pm
            resultV.append(pr)
            resultT.append(i)
    print(resultT[-1]+1)
    return resultT,resultV,(resultV.index(max(resultV))+1+1)+sumv
x,y,p = price_game(sumv=500,v=1,p=0.01)
plt.plot(y,x),p
```

```python
def extremum(srcData,deltaValue,probabilityCurve):
    if isinstance(probabilityCurve,(int,float)):
        deltaProb = np.cumsum(np.array([probabilityCurve]*(1/probabilityCurve).__int__()), dtype=float)
    if isinstance(deltaValue,(int,float)):
        deltaV = np.cumsum(np.array([deltaValue]*len(deltaProb).__int__()), dtype=float)
    if isinstance(probabilityCurve,(np.ndarray)):
        deltaProb = probabilityCurve
    if isinstance(probabilityCurve,(np.ndarray)):
        deltaV = deltaValue
    dpn = deltaProb.shape
    dvn = deltaV.shape
    if dpn==dvn:
        decision_chain = deltaV*(1-deltaProb)
        return srcData+np.where(decision_chain==decision_chain.max())[0]+1,deltaProb,deltaV,decision_chain

P,deltaProb,deltaV,decision_chain = extremum(500,1,0.01)
plt.plot(decision_chain)
P
```
