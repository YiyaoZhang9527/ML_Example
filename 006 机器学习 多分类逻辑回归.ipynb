# OVO的方式多分类逻辑回归做拟合

```python
import numpy as np
from sklearn.linear_model import LogisticRegression
import matplotlib.pyplot as plt


def lr(trainX,trainY,TestX):
    log_reg1=LogisticRegression(multi_class="multinomial",solver='newton-cg')
    log_reg1.fit(trainX,trainY)
    #display(log_reg1.intercept_)  #截距
    #display(log_reg1.coef_)  #线性模型的系数
    return log_reg1.predict(TestX),log_reg1.intercept_,log_reg1.coef_
```

```python
x = np.random.rand(10000).reshape(1000,10)
y = (np.random.randn(1000)*10).astype(int)#<x.mean()).astype(int)
train_x ,train_y , test_x ,test_y = x[500:],y[500:],x[:500],y[:500]
```

```python
predict = lr(train_x,train_y,test_x)
```

```python
plt.plot(predict[0]-test_y)
```
