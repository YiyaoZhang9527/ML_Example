## 加法链式法则

```python
X = np.linspace(0,np.pi*2,1000)
```

```python
x = sympy.Symbol('x')
```

```python
f = sympy.sin(x)+x**2
f
```

```python
df = sympy.diff(f,x)
df
```

## 带入值画图看看

```python
dfline = [df.subs(x,i).evalf(7) for i in range(-10,10)]
```

```python
sinline = np.sin(np.arange(-10,10))
xxline = np.power(np.arange(-10,10),2)
```

```python
plt.figure(figsize=(10,8))
plt.plot(dfline,label='fx(x)=x^2+sin(x) 的导数函数 df(x)',color='r')
plt.plot(sinline,label='sin(x)',color='g')
plt.plot(xxline,label='x^2',color='b')
plt.xlabel('index_of_X')
plt.ylabel('index_of_Y')
plt.plot()
plt.show()
plt.close()
```

## 链式乘法法则

```python
f1 = sympy.sin(x)*(x**2)
f1
```

## 求个导

```python
df2 = sympy.diff(f1)
df2
```

```python
sinline2 = np.sin(np.arange(-100,100))
xxline2 = np.power(np.arange(-100,100),2)
xindex = np.zeros(xxline.shape)
```
