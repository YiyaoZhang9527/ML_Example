```python
import scipy
import matplotlib.pyplot as plt
import numpy as np
```

```python
line = np.linspace(1,100,1000)
```

```python
plt.plot(line)
```

```python
plt.plot(line*2)
plt.plot(line*3)
plt.plot(line*4)
plt.plot(line*5)
plt.plot(line*6)
plt.plot(line*7)
plt.plot(line*8)
plt.plot(line*9)
plt.plot(line*10)
plt.plot(line*11)
```

```python
plt.plot(line**2)
plt.plot(line**3)
```

```python
plt.plot(np.log(line))
plt.plot(np.log2(line))
plt.plot(np.log10(line))
```

```python
plt.plot(np.sin(line))
plt.plot(np.cos(line))
plt.plot(np.sin(np.linspace(0,np.pi*2,1000)))
plt.plot(np.cos(np.linspace(0,np.pi*2,1000)))
```

```python
plt.plot(np.sin(np.linspace(0,np.pi,100)))
```

```python
plt.figure(figsize=(16,9))
pointsfactorial = [np.math.factorial(i) for i in range(10)]
plt.plot(pointsfactorial)
plt.scatter(range(len(pointsfactorial)),pointsfactorial)
```
