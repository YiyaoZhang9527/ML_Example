```python
from PIL import Image as pl
import numpy as np
x = np.array([[1, 2, 3], [4, 5, 6]])
y = pl.new("RGB", (104*8, 104*8), color=0)
import IPython.display
import os
os.getcwd()
```

```python
f = open("洪水填充数据.txt",'r').read()
cols =  eval(f)
```

```python
basex = 3704 
basey = 5624
insite = (48, 52)
cols[insite[0]][insite[1]]
BLOCK_MOVEMENT_NORTH_WEST = 0x1;
BLOCK_MOVEMENT_NORTH = 0x2;
BLOCK_MOVEMENT_NORTH_EAST = 0x4;
BLOCK_MOVEMENT_EAST = 0x8;
BLOCK_MOVEMENT_SOUTH_EAST = 0x10;
BLOCK_MOVEMENT_SOUTH = 0x20;
BLOCK_MOVEMENT_SOUTH_WEST = 0x40;
BLOCK_MOVEMENT_WEST = 0x80;
BLOCK_MOVEMENT_OBJECT = 0x100;
BLOCK_MOVEMENT_FLOOR_DECORATION = 0x40000;
BLOCK_MOVEMENT_FLOOR = 0x200000; # Eg. water
BLOCK_MOVEMENT_FULL = BLOCK_MOVEMENT_OBJECT | BLOCK_MOVEMENT_FLOOR_DECORATION | BLOCK_MOVEMENT_FLOOR;
def paintsq(i, j, color):
    for k in range(0,8):
        for l in range(0,8):
            y.putpixel((i*8+k,j*8+l), color)
for i in range(0,100):
    for j in range(0,100):
        if ((BLOCK_MOVEMENT_WEST & cols[i][j]) != 0):
            for k in range(0,8):
                y.putpixel((i*8+0,j*8+k), 255 + 255 * 256 + 256*256*255)
        if ((BLOCK_MOVEMENT_EAST & cols[i][j]) != 0):
            for k in range(0,8):
                y.putpixel((i*8+7,j*8+k), 255 + 255 * 256 + 256*256*255)
        if ((BLOCK_MOVEMENT_NORTH & cols[i][j]) != 0):
            for k in range(0,8):
                y.putpixel((i*8+k,j*8+7), 255 + 255 * 256 + 256*256*255)
        if ((BLOCK_MOVEMENT_SOUTH & cols[i][j]) != 0):
            for k in range(0,8):
                y.putpixel((i*8+k,j*8+0), 255 + 255 * 256 + 256*256*255)
        if ((BLOCK_MOVEMENT_FLOOR & cols[i][j]) != 0):
            paintsq(i,j, 255)
        if ((BLOCK_MOVEMENT_FLOOR_DECORATION & cols[i][j]) != 0):
            paintsq(i,j, 255)
        if (((BLOCK_MOVEMENT_OBJECT & cols[i][j]) != 0) & ((BLOCK_MOVEMENT_FLOOR & cols[i][j]) == 0)):
            paintsq(i,j, 255+ (255 << 8))
IPython.display.display(y)
```
