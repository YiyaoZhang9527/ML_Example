## 理解隐藏层的激活函数，激活函数是将信号归于0～1之间，来做判定依据

```python
def sigmoid(x):
  '''这里我们用的是逻辑斯蒂曲线'''
  return 1/(1+np.exp(-x))
```

```python
plt.figure(figsize=(10,10))
plt.plot(np.linspace(-10,10,100))
```

## 当我们用逻辑斯蒂函数作用于它后，它将在0到1之间变换，从而让我们得到一个预测结果的权重概率

```python
plt.figure(figsize=(10,10))
plt.plot(sigmoid(np.linspace(-10,10,100)))
```

## 1.1输入层到第一层的第一个神经元的传递

```python
X = np.array([1.0,0.5]) #一维数组
X
```

#### 设置权重,权重的个数要与输入层的元素保持一致

```python
W1 = np.array([[0.1,0.3,0.5],[0.2,0.4,0.6]])
W1
```

#### 隐藏层的加权和

```python
B1 = np.array([0.1,0.2,0.3])
A1 = np.dot(X,W1) + B1
('是隐藏层的加权和:',A1)
```

#### 是经过激活函数数字转化后的信号

```python
Z1 = sigmoid(A1)
Z1
```

## 1.2第一层到第二层到的神经元传递

```python
W2 = np.array([[0.1,0.3],[0.2,0.4],[0.3,0.6]])
B2 = np.array([0.1,0.2])
A2 = np.dot(Z1,W2) + B2
Z2 = sigmoid(A2)
Z2
```

## 1.3随意再加一层？第二层到第三层的神经元传递开始，哈哈

```python
W3 = np.array([[0.1,0.3],[0.2,0.4]])
B3 = np.array([0.1,0.2])
A3 = np.dot(Z2,W3) + B3
Z3 = sigmoid(A3)
Z3
```

# 一个完整的样例
其实就是把上面的代码过程整理一下

```python
def sigmoid(x):
  '''这里我们用的是逻辑斯蒂曲线'''
  return 1/(1+np.exp(-x))

def identity_function(x):
  '''可以调整，获得更好的精度'''
  return x

def init_network():
  network = dict()
  network["W1"] = np.array([[0.1,0.3,0.5],[0.2,0.4,0.6]]) #第一层权重
  network["b1"] = np.array([0.1,0.2,0.3]) #第一层偏置
  network["W2"] = np.array([[0.1,0.3],[0.2,0.4],[0.3,0.6]]) #第二层权重
  network["b2"] = np.array([0.1,0.2]) #第二层偏置
  network["W3"] = np.array([[0.1,0.3],[0.2,0.4]]) #第三层权重
  network["b3"] = np.array([0.1,0.2]) #第三层偏置
  return network

def forward(network,x):
  W1, W2 ,W3 = network["W1"],network["W2"],network["W3"]
  b1 ,b2 ,b3 = network["b1"],network["b2"],network["b3"]
  Z_func = lambda x,w,b : sigmoid(np.dot(x,w)+b)
  z1 = Z_func(x,W1,b1)
  z2 = Z_func(z1,W2,b2)
  a3 = np.dot(z2,W3)+b3
  y = identity_function(a3)
  return y

network = init_network()
x = np.array([1.0,0.5])
y = forward(network,x)
y
```
