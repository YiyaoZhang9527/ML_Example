```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
points = np.linspace(0,np.pi**2,100)

x = np.cos(points)
y = np.sin(points)
plt.figure(figsize=(10,10))
plt.plot(x,y)
```

```python
xpoint = np.random.random_integers(0,10,10)
ypoint = np.random.random_integers(0,10,10)
plt.plot(xpoint,ypoint)
plt.scatter(xpoint,ypoint,c='r')
```

```python
xpoint,ypoint
```

```python
address = np.c_[xpoint,ypoint]
address
```

```python
uninx = np.linspace(address[0][0],address[-1][0],100)
uniny = np.linspace(address[0][-1],address[-1][-1],100)
```

```python
uninx,uniny
```

```python
plt.plot(uninx,uniny)
plt.plot(np.cos(np.linspace(0,np.pi,100))*uninx,np.cosh(np.linspace(0,np.pi,100))*uniny)
```

```python
plt.plot(uninx,uniny,'g')
plt.plot(xpoint,ypoint,'black')
plt.scatter(xpoint,ypoint,c='r')
plt.plot(np.cos(np.linspace(0,np.pi,100))*uninx,np.cosh(np.linspace(0,np.pi,100))*uniny,c='y')
```
