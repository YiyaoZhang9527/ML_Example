```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
plt.rcParams['font.sans-serif'] = ['SimHei']  # 解决中文显示问题-设置字体为黑体
plt.rcParams['axes.unicode_minus'] = False  # 解决保存图像是负号'-'显示为方块的问题
```

```python
line = np.linspace(0,10,100)
plt.figure(figsize=(16,9))
plt.title("等差序列线")
plt.plot(line,c='r')
plt.scatter(x=range(len(line)),y=line,c='g')
plt.show()
plt.close()
```

```python
plt.figure(figsize=(16,9))
plt.title("负指数线")
plt.plot(np.exp(-line),c='r')
plt.scatter(x=range(len(line)),y=np.exp(-line),c='g')
plt.show()
plt.close()
```
