```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def fx1(x):
    return 0.3*(x**2)

data = fx1(np.linspace(0,1,10000))
plt.figure(figsize=(10,10))
plt.plot(data)
```

### 完全公式计算导数  k = (y1-y2)/(x1-x2)

```python
def dx1(x):
    return [(x[n]-x[n+1])/(n-(n+1)) for n in range(len(x)-1)]

plt.plot(dx1(data))
```

### 因为x增量等于-1，相除后又负负得正， 简化后等效与 k = y2-y1

```python
def dx2(x):
    return [(x[n+1]-x[n]) for n in range(len(x)-1)]

plt.plot(dx2(data))
```
